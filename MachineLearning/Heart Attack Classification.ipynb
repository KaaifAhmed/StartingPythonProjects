```python
df = pd.read_csv('heart.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_new = df.sample(frac=1, random_state=1)
```

```python
303*0.7
```

```python
train = df_new[:212].reset_index(drop=True)
test = df_new[212:].reset_index(drop=True)
```

```python
train
```

```python
test
```

```python
df_new.columns
```

```python
x_train = train[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh',
       'exng', 'oldpeak', 'slp', 'caa', 'thall']]
y_train = train['output']
```

```python
x_test = test[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh',
       'exng', 'oldpeak', 'slp', 'caa', 'thall']]
y_test = test['output']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
prediction
```

```python
y_test
```

```python
pd.DataFrame({'Original':y_test, 'Prediction':prediction})
```

```python
model.score(x_test, y_test)
```

```python
model = LogisticRegression(max_iter=57, C=0.6)
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

```python
pd.DataFrame({'Original':y_test, 'Predicted':prediction})
```

```python
model.score(x_test, y_test)
```
